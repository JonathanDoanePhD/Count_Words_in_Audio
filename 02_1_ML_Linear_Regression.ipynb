## Models

### 1. Linear Regression



```python
from sklearn.linear_model import LinearRegression
from scipy import stats as scst
```

#### a. Let's use nearly all 2,000 observations

```python
print(gm30)

TR_num_onsets_df = TR_df.copy()
TR_num_onsets_df['median_func_num'] = gm30
TR_num_onsets_df = TR_num_onsets_df.dropna()

X = TR_num_onsets_df.loc[:,'num_onsets'].to_list()
y = TR_num_onsets_df.loc[:,'median_func_num'].to_list()

linr = scst.linregress(X, y)

def lin_func(x):
    return linr[0]*x + linr[1]

slr = list(map(lin_func, X))

plt.scatter(X, y)
plt.plot(X, slr, color='r')
plt.xlabel('num_onsets')
plt.ylabel('median_func_num')
plt.title('y = '+str(round(linr[0],2))+'*x + '+str(round(linr[1],2))+',  Correlation Coefficient r = '+str(linr[2]))
plt.rcParams["figure.figsize"] = my_figsize
plt.show()
```

#### b. Let's use my_thresh = 0.85, so 1,151 observations

```python
print(gm85)

TR_num_onsets_85_df = TR_df.copy()
TR_num_onsets_85_df['median_func_num'] = gm85
TR_num_onsets_85_df = TR_num_onsets_85_df.dropna()

X = TR_num_onsets_85_df.loc[:,'num_onsets'].to_list()
y = TR_num_onsets_85_df.loc[:,'median_func_num'].to_list()

linr = scst.linregress(X, y)

def lin_func(x):
    return linr[0]*x + linr[1]

slr = list(map(lin_func, X))

plt.scatter(X, y)
plt.plot(X, slr, color='r')
plt.xlabel('num_onsets')
plt.ylabel('median_func_num')
plt.title('y = '+str(round(linr[0],2))+'*x + '+str(round(linr[1],2))+',  Correlation Coefficient r = '+str(linr[2]))
plt.rcParams["figure.figsize"] = my_figsize
plt.show()
```

#### c. Find the strongest r

```python
R=[]
for t in range(99):
    t = (t+1)/100
    t_gm = get_meds(t,my_df)
    t_df = TR_df.copy()
    t_df['median_func_num'] = t_gm
    t_df = t_df.dropna()
    t_X = t_df.loc[:,'num_onsets'].to_list()
    t_y = t_df.loc[:,'median_func_num'].to_list()
    t_r = scst.linregress(t_X, t_y)[2]
    R.append(t_r)

X = range(1,100)
y = R

plt.plot(X, y)
plt.xlabel('r')
plt.ylabel('Threshold')
plt.title('Correlation Coefficients r')
plt.rcParams["figure.figsize"] = my_figsize
plt.show()
```

### d. Two last tests; 0.83 and 0.99

```python
for t in [.83,.99]:
    t_gm = get_meds(t,my_df)
    t_df = TR_df.copy()
    t_df['median_func_num'] = t_gm
    t_df = t_df.dropna()
    t_X = t_df.loc[:,'num_onsets'].to_list()
    t_y = t_df.loc[:,'median_func_num'].to_list()
    t_r = scst.linregress(t_X, t_y)[2]
    #
    t_linr = scst.linregress(t_X, t_y)
    def t_lin_func(x):
        return t_linr[0]*x + t_linr[1]
    t_slr = list(map(t_lin_func, t_X))
    plt.scatter(t_X, t_y)
    plt.plot(t_X, t_slr, color='r')
    plt.xlabel('num_onsets')
    plt.ylabel('median_func_num')
    plt.title('y = '+str(round(t_linr[0],2))+'*x + '+str(round(t_linr[1],2))+',  Correlation Coefficient r = '+str(t_linr[2]))
    plt.rcParams["figure.figsize"] = my_figsize
    plt.show()
    print('\n'*2)
```
