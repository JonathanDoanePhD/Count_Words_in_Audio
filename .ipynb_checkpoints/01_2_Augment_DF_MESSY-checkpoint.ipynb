```python
TR_dir = r"commonvoice\train\clips"
TR_aud = os.listdir(TR_dir)
TE_dir = r"commonvoice\test\clips"
TE_aud = os.listdir(TE_dir)
VA_dir = r"commonvoice\validation\clips"
VA_aud = os.listdir(VA_dir)

TTV_dir = [TR_dir, TE_dir, VA_dir]
TTV_aud = [TR_aud, TE_aud, VA_aud]

# data, samplerate = librosa.load(TR_dir+"\\"+TR_aud[0])
# length = len(data)/samplerate

for j in range(2):
    j=j+1
    dom_freqs = np.zeros(len(TTV_aud[j]))
    # iterate over all recordings in training set
    for i in range(len(TTV_aud[j])):
        ## load the recording
        data, samplerate = lr.load(TTV_dir[j]+"\\"+TTV_aud[j][i])
        length = len(data)/samplerate
        ## fourier transform
        freqdata = scipy.fft.rfft(data)
        timestep = length/data.shape[0]
        xf = scipy.fft.rfftfreq(data.shape[0], d=timestep)
        ## limit frequency range from 0Hz - 500 Hz
        # note entries of xf are evenly spaced
        diff = xf[1]-xf[0]
        # numbr of entries of xf to keep 
        maxfreqrange = 500
        numentries = round(maxfreqrange/diff)
        # limit searching for maxing amplitude within this range
        xf = xf[0:numentries+1]
        freqdata = freqdata[0:numentries+1]
        dom_freqs[i] = xf[np.argmax(freqdata)]
        # print(i)
    
    TTV_df[j]['dom_freq'] = dom_freqs
```

```python
TE_MedMaker_df = TE_df.iloc[:,76:140].copy()
VA_MedMaker_df = VA_df.iloc[:,76:140].copy()

my_thresh = .85
my_df = TE_MedMaker_df.copy()

def get_meds(in_thresh=my_thresh, in_df=my_df):
    t_rows = len(in_df.iloc[:,0].to_list())
    out_meds = []
    for i in range(t_rows):
        t_list = in_df.iloc[i,:].to_list()
        t_indices = [index for index, elt in enumerate(t_list) if in_thresh <= elt and elt <= 1/in_thresh]
        if len(t_indices) == 0:
            out_meds.append(None)
        else:
            out_meds.append( int(st.median(t_indices)) )
    return out_meds

TE_gm30 = get_meds(.30, TE_MedMaker_df)
TE_gm85 = get_meds(.85, TE_MedMaker_df)
TE_gm99 = get_meds(.99, TE_MedMaker_df)

VA_gm30 = get_meds(.30, VA_MedMaker_df)
VA_gm85 = get_meds(.85, VA_MedMaker_df)
VA_gm99 = get_meds(.99, VA_MedMaker_df)
```

```python
TE_df['median_func_num'] = TE_gm85

VA_df['median_func_num'] = VA_gm85
```

```python
def get_MEDS(in_df=my_df):
    gm = []
    # Collect medians in desirable threshold ranges, but eventually soften the threshold to include every person
    for t in reversed(range(5,100,5)):
        t = t/100
        gm.append(get_meds(t,in_df))
    # Get rid of all None values (in the most lenient threshold); I placed the median of [0:63] basically.
    l = gm[-1]
    while None in l:
        l[l.index(None)] = 31
    # Collect the best medians possible
    gM = []
    for row in range(len(in_df.iloc[:,0].to_list())):
        i = 0
        j = gm[i][row]
        while None == j:
            i=i+1
            j = gm[i][row]
        gM.append(j)
    return gM
```
